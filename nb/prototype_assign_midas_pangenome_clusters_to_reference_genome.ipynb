```python
midas_gene_list = idxwhere(strain_scores.strain_depth > 0)

_correlation = strain_scores.strain_corr[midas_gene_list]
_depth = strain_scores.strain_depth[midas_gene_list]
_bitscore = orf_x_midas.unstack().reindex(columns=midas_gene_list, fill_value=0).astype(float)

for c_thresh in [0, 0.25, 0.5, 0.75, 0.8, 0.9, 0.95, 0.99, 0.995]:
    d_thresh = 0.25
    b_thresh = 0.5
    _bitscore_hit = _bitscore > b_thresh
    _spgc_hit = (_correlation > c_thresh) & (_depth > d_thresh)

    _match = (_spgc_hit & _bitscore_hit)

    tp_midas = idxwhere(_bitscore_hit.loc[:, _spgc_hit].any(axis=0))  # MIDAS genes hit by both.
    fp_midas = idxwhere(~(_bitscore_hit.loc[:, _spgc_hit].any()))  # MIDAS genes were hit by SPGC but never by BLAST?
    fn_orf = idxwhere(~(_bitscore_hit.loc[:, _spgc_hit].any(axis=1)))  # How many ORFs were hit by BLAST but no matching SPGC hits?
    tp_orf = idxwhere((_spgc_hit & _bitscore_hit).any(axis=1))  # How many ORFs were hit by BLAST and by SPGC?
    
    n_tp_midas = len(tp_midas)
    n_fp_midas = len(fp_midas)
    n_fn_orf = len(fn_orf)
    n_tp_orf = len(tp_orf)

    precision_midas = n_tp_midas / (n_tp_midas + n_fp_midas)
    recall_orf = n_tp_orf / (n_tp_orf + n_fn_orf)

    print(f'{c_thresh}\t{d_thresh}\t{precision_midas:0.2f}\t{recall_orf:0.2f}\t{n_fp_midas}\t{n_fn_orf}\t{n_tp_midas}\t{n_tp_orf}')

# MIDAS genes that map to multiple ORFs, tend to have higher depth ratios.
#sns.regplot(x=_bitscore_hit.loc[tp_orf, tp_midas].sum(axis=0), y=strain_scores.loc[tp_midas].strain_depth)
```

```python
from itertools import product

midas_gene_list = idxwhere((strain_scores.strain_corr > 0) | (strain_scores.strain_depth > 0))

_correlation = strain_scores.strain_corr[midas_gene_list]
_depth = strain_scores.strain_depth[midas_gene_list]
_bitscore = orf_x_midas.unstack().reindex(columns=midas_gene_list, fill_value=0).astype(float)

c_thresh_list_complement = np.logspace(-2, 0, num=10)
c_thresh_list = list(reversed(1 - c_thresh_list_complement))
d_thresh_list = np.linspace(0, 1.0, num=10)

b_thresh = 0.8
_bitscore_hit = _bitscore > b_thresh

hits_matching = np.empty((len(d_thresh_list), len(c_thresh_list)))
orfs_matching = np.empty_like(hits_matching)
for (i, d_thresh), (j, c_thresh) in tqdm(product(enumerate(d_thresh_list), enumerate(c_thresh_list))):
    _hits = (_correlation > c_thresh) & (_depth > d_thresh)
    # hits_matching[i, j], orfs_matching[i, j] = two_way_accuracy(_bitscore_hit, _hits  # FIXME:
    hits_matching[i, j] = (_bitscore_hit).any(axis=0)[_hits].mean()
    orfs_matching[i, j] = (_bitscore_hit & _hits).any(axis=1).mean()
    
    
fig, axs = plt.subplots(3, figsize=(5, 10))

ax = axs[0]
ax.set_title('hits_matching')
cax = make_axes_locatable(ax).append_axes('right', size='5%', pad=0.05)
artist = ax.pcolormesh(c_thresh_list_complement, d_thresh_list, hits_matching, norm=mpl.colors.PowerNorm(1, vmin=0, vmax=1))
ax.set_xscale('log')
# ax = plt.gca()
# ax.invert_xaxis()
ax.set_xlim(right=1)
# ax.set_ylim(0, 1)
cbar = fig.colorbar(artist, cax=cax)

ax = axs[1]
ax.set_title('orfs_matching')
cax = make_axes_locatable(ax).append_axes('right', size='5%', pad=0.05)
artist = ax.pcolormesh(c_thresh_list_complement, d_thresh_list, orfs_matching, norm=mpl.colors.PowerNorm(1, vmin=0, vmax=1))
ax.set_xscale('log')
ax.set_xlim(right=1)
# ax.set_ylim(0, 1)
cbar = fig.colorbar(artist, cax=cax)

ax = axs[2]
ax.set_title('product')
cax = make_axes_locatable(ax).append_axes('right', size='5%', pad=0.05)
artist = ax.pcolormesh(c_thresh_list_complement, d_thresh_list, hits_matching * orfs_matching, norm=mpl.colors.PowerNorm(1, vmin=0, vmax=1))
ax.set_xscale('log')
ax.set_xlim(right=1)
# ax.set_ylim(0, 1)
cbar = fig.colorbar(artist, cax=cax)

fig.tight_layout()
```
